#### 2.	Import both the attached files (student.csv and marks.csv) in Jupyter notebook and perform following operations:

```python
student_df = pd.read_csv('student.csv')
marks_df = pd.read_csv('marks.csv')

combined_df = pd.merge(student_df, marks_df, on='Student_id', how='outer')


print(combined_df)
```

```python
genderwise_marks = combined_df.groupby('Gender')['Mark'].agg(['max', 'min'])

print(genderwise_marks)
```

```python
average = combined_df['Mark'].mean()
print (average)
```

```python
more_than_average = combined_df[combined_df['Mark'] > average]
print (more_than_average[['Student_id', 'Mark']])
```

```python
employed_males = combined_df[(combined_df['Gender'] == 'Male') & (combined_df['Employed'] == 'yes')]

print(employed_males)
```

```python

# Function to categorise IQ level

def categorize_iq(Mark):
    if Mark > 80:
        return 'Intelligent'
    elif 50 <= Mark <= 80:
        return 'Mediocre'
    else:
        return 'Weak'

# Apply the function to create the 'IQ_level' column
combined_df['IQ_level'] = combined_df['Mark'].apply(categorize_iq)

print(combined_df)
```

```python
gender_count = combined_df.groupby(['City', 'Gender']).size().unstack(fill_value=0)

print(gender_count)
```

```python
# Filter the DataFrame to only include male students
male_students = combined_df[combined_df['Gender'] == 'Male']

# Sort the male students by score in descending order
top_5_male_scorers = male_students.sort_values(by='Mark', ascending=False).head(5)

print(top_5_male_scorers[['Student_id', 'Mark']])
```

```python
# Replace 'Male' with 'M' and 'Female' with 'F'
combined_df['Gender'] = combined_df['Gender'].replace({'Male': 'M', 'Female': 'F'})

# Export to CSV file at D: drive
file_path = r'D:\test.csv'
combined_df.to_csv(file_path, index=False)

print(f"Data exported to {file_path}")
```

```python
duplicated_student_ids = combined_df['Student_id'].duplicated().any()

print('duplicated_student_ids =', duplicated_student_ids)
```

```python
# Create a new DataFrame with only the integer/float columns
numeric_df = combined_df.select_dtypes(include=['number'])

print(numeric_df)
```
